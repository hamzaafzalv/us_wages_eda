# Data Cleaning

```python
df.isnull().sum() #checking number of null values
```

```python
df_clean = df.drop
```

```python
df_clean =df.drop(columns=['WorldRegion'])
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.isnull().sum()
```

```python
df_clean.info()
```

```python
df_clean #checking if the data is in correct shape relative to it's column name
```

```python
df_clean = df_clean.set_index('RecordNumber')
df_clean
```

```python
df_clean['ZipCodeType'].value_counts()
```

```python
df_clean = df_clean [df_clean ['ZipCodeType'] == 'STANDARD']
df_clean #removed PO BOX zipcodes because they are not physical address
```

```python
df_clean = df_clean.drop(columns='Location')
df_clean #dropped location Column because it was almost a duplicate of LocationText column
```

```python
df_clean 
```

```python
df_clean.rename(columns={'LocationText': 'Location'}, inplace=True)
df_clean
```

# Data Visualization

```python
wages_corr= df_clean[['Zipcode','TotalWages','TaxReturnsFiled','Lat','Long']].corr(method='pearson')
plt.figure(figsize=(14,5))
heatmap = sns.heatmap(wages_corr, annot=True)
heatmap.set_title('Correlation between Wages and Zipcode')
```

```python
df_clean['LocationType'].value_counts()
df_clean['Country'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df_clean,y='TotalWages',x='Lat',color='c',marker='*').set(title='Wages Graph from South to North of US')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df_clean,y='TotalWages',x='Long',color='c',marker='*').set(title='Wages Graph from West to East of US')
plt.show()
```
